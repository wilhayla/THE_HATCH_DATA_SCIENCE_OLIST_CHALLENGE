### LIMPIEZA DEL DATAFRAME (**df_order_payments**)

```python
# Filtrar las filas que tienen en la columna 'payment_value' valor igual a cero.
# Utilizar la tecnica de Indexacion Booleana (Boolean Indexing) para crear una mascara que identifique y seleccione solo esas filas.

# 1. Definir la Condicion (la Mascara Booleana)
# Creamos una Serie de Pandas de True/False donde True indica que payment_value es CERO.
mask_cero_payment = (df_order_payments['payment_value'] == 0.00)

# 2. Aplicar la mascara para filtrar.
# Seleccionamos solo las filas donde la mascara es True.
filas_con_pago_cero = df_order_payments[mask_cero_payment]

# 3. Mostrar resultado.
print(f"Numero de pagos con valor cero: {filas_con_pago_cero.shape[0]}")
filas_con_pago_cero.head(10)
```

```python
col_cat = 'payment_type'
fig, ax = plt.subplots(figsize=(8,5))

sns.countplot(x=col_cat, data=df_order_payments, ax=ax)

ax.set_title(f'Distribución de Valores en {col_cat}', fontsize=14)
ax.set_xlabel(col_cat, fontsize=12)
ax.set_ylabel('Conteo', fontsize=12)

ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right')

plt.show
```

```python
# Filtrar y contar la categoria not_defined.
# 1. Filtrar los pagos "no definidos"

df_not_defined = df_order_payments[df_order_payments['payment_type'] == 'not_defined']

print(f"Número total de pagos 'not_defined': {df_not_defined.shape[0]}")
print("\nEstadísticas de 'payment_value' para 'not_defined':")
print(df_not_defined['payment_value'].describe())

# 2. Verificar si todos los pagos 'not defined' son de valor cero.

zero_value_in_not_defined = df_not_defined[df_not_defined['payment_value'] == 0.00].shape[0]
print(f"\nNúmero de pagos 'not_defined' con valor CERO: {zero_value_in_not_defined}")
```

```python
# LIMPIEZA FINAL.
# Actualizar el dataframe con uno nuevo con solo los valores que son mayor a cero.
df_order_payments_clean = df_order_payments[df_order_payments['payment_value'] > 0.00].copy()

print(f"Filas originales: {df_order_payments.shape[0]}")
print(f"Filas después de la limpieza (pagos > 0): {df_order_payments_clean.shape[0]}")
```
