+ ### **Limpieza** del Data Frame Order Payments (**df_order_payments**)

```python
# Filtrar las filas que tienen en la columna 'payment_value' valor igual a cero.
# Utilizar la tecnica de Indexacion Booleana (Boolean Indexing) para crear una mascara que identifique y seleccione solo esas filas.

# 1. Definir la Condicion (la Mascara Booleana)
# Creamos una Serie de Pandas de True/False donde True indica que payment_value es CERO.
mask_cero_payment = (df_order_payments['payment_value'] == 0.00)

# 2. Aplicar la mascara para filtrar.
# Seleccionamos solo las filas donde la mascara es True.
filas_con_pago_cero = df_order_payments[mask_cero_payment]

# 3. Mostrar resultado.
print(f"Numero de pagos con valor cero: {filas_con_pago_cero.shape[0]}")
filas_con_pago_cero.head(10)
```

```python
# LIMPIEZA FINAL.
# Actualizar el dataframe con uno nuevo con solo los valores que son mayor a cero.
df_order_payments_clean = df_order_payments[df_order_payments['payment_value'] > 0.00].copy()

print(f"Filas originales: {df_order_payments.shape[0]}")
print(f"Filas despuÃ©s de la limpieza (pagos > 0): {df_order_payments_clean.shape[0]}")
```
